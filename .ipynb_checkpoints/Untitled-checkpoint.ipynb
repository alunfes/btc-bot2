```python
time_kijun = 3600
loss_kijun = 1500
profit_kijun = 5000

ok_points = {} #index,buy or sell

for i in range(len(tick) - time_kijun):
    kijun_p = tick[i]
    buy_judge = 0
    sell_judge = 0
    for j in range(time_kijun):
        if tick[i+j] - kijun_p <= loss_kijun:
            buy_judge = -1
        elif tick[i+j] - kijun_p >= loss_kijun:
            sell_judge = -1
        elif buy_judge >=0 and  tick[i+j] - kijun_p >= profit_kijun:
            ok_points[i] = 'buy'
            break
        elif buy_judge >=0 and  kijun_p - tick[i+j] >= profit_kijun:
            ok_points[i] = 'sell'
            break
        elif buy_judge == -1 and sell_judge == -1:
            break
    
    
    
    
```

```python
import numpy as np

tick = []
split_num = 60
for i in range(len(md.ohlc_bot.dt)):
    sec_width = 0
    om_tick = []
    if md.ohlc_bot.open[i] < md.ohlc_bot.close[i]: #open, low, high, close
        ol = md.ohlc_bot.open[i] - md.ohlc_bot.low[i]
        lh = md.ohlc_bot.high[i] - md.ohlc_bot.low[i]
        hc = md.ohlc_bot.high[i] - md.ohlc_bot.close[i]
        sec_width = (ol + lh + hc) / split_num
        om_tick.extend(list(np.round(np.linspace(md.ohlc_bot.open[i], md.ohlc_bot.low[i],round(ol/sec_width)))))
        om_tick.extend(list(np.round(np.linspace(md.ohlc_bot.low[i], md.ohlc_bot.high[i],round(lh/sec_width)))))
        om_tick.extend(list(np.round(np.linspace(md.ohlc_bot.high[i], md.ohlc_bot.close[i],round(hc/sec_width)))))
    else: #open, high, low, close
        oh = md.ohlc_bot.high[i] - md.ohlc_bot.open[i]
        hl = md.ohlc_bot.high[i] - md.ohlc_bot.low[i]
        lc = md.ohlc_bot.close[i] - md.ohlc_bot.low[i]
        sec_width = (oh + hl + lc) / split_num
        om_tick.extend(list(np.round(np.linspace(md.ohlc_bot.open[i], md.ohlc_bot.high[i],round(oh/sec_width)))))
        om_tick.extend(list(np.round(np.linspace(md.ohlc_bot.high[i], md.ohlc_bot.low[i],round(hl/sec_width)))))
        om_tick.extend(list(np.round(np.linspace(md.ohlc_bot.low[i], md.ohlc_bot.close[i],round(lc/sec_width)))))
    tick.extend(om_tick)
```

```python
def check_matched_ind_with_train_x(train_x):
    ary = list(train_x['ma_kairi5'])[0:30]
    key = 5
    for i in range(len(MarketData.ohlc_data.ma_kairi[key])):
        flg = True
        for j  in range(len(ary)):
            if ary[j] != MarketData.ohlc_data.ma_kairi[key][i+j]:
                flg = False
        if flg:
            return i
    return -1

check_matched_ind_with_train_x(test_x)
```

```python
import time
from MarketData2 import MarketData2
from CryptowatchDataGetter import CryptowatchDataGetter
from XgbModel import XgbModel
from datetime import datetime
import xgboost as xgb
import copy

print('bot - initializing MarketData2..')
MarketData2.initialize_from_bot_csv(100, 30, 500)
train_df = MarketData2.generate_df(MarketData2.ohlc_bot)
model = XgbModel()
print('bot - training xgb model..')
train_x, test_x, train_y, test_y = model.generate_data(copy.deepcopy(train_df), 1)
bst = model.train(train_x, train_y)
print('bot - training completed..')
print('bot - updating crypto data..')
MarketData2.initialize_from_bot_csv(100, 30, 500)
MarketData2.ohlc_bot.cut_data(200)
while True:
    if datetime.now().second == 1:
        res, omd = CryptowatchDataGetter.get_data_after_specific_ut(MarketData2.ohlc_bot.unix_time[-1])
        if res == 0:
            start = time.time()
            for i in range(len(omd.dt)):
                MarketData2.ohlc_bot.add_and_pop(omd.unix_time[i],omd.dt[i], omd.open[i], omd.high[i], omd.low[i], omd.close[i], omd.size[i])
            MarketData2.update_ohlc_index_for_bot2()
            df = MarketData2.generate_df_for_bot(copy.deepcopy(MarketData2.ohlc_bot))
            pred_x = model.generate_bot_pred_data(df)
            predict = bst.predict(xgb.DMatrix(pred_x))
            elapsed_time = time.time() - start
            print("elapsed_time:{0}".format(elapsed_time) + "[sec]")
            print('dt={}, open={}, close={}, predict={}'.format(MarketData2.ohlc_bot.dt[-1],MarketData2.ohlc_bot.open[-1],
                                                                MarketData2.ohlc_bot.close[-1], predict[0]))
        else:
            print('crypto watch data download error!')
```
